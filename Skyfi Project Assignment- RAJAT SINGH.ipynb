```python
X = df['LSTAT'].values
Y = df['target'].values

print(Y[ : 5])   # before performing noamalization
```

```python
x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1, 1))
X = X[:, -1]
y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1, 1))
Y = Y[:, -1]
print(Y[:5])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2)
```

```python
def error(m, x, c, t):
    n = x.size
    pred = m * x + c
    e = sum((pred-t)**2)
    return e*(1/(2*n))
```

```python
def update(m, x, c, t, learning_rate):
    pred = m*x + c
    grad_m = sum(2*(pred-t)*x) * (1/len(x))
    grad_c = sum(2*(pred-t)) * (1/len(x))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m, c
```

```python
def gradient(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = list()
    mc_values = list()
    for i in range (iterations):
        e = error(m, x, c, t)
        if e < error_threshold:
            print('Error less than threshold')
            break
        error_values.append(e)
        m, c = update(m, x, c, t, learning_rate)
        mc_values.append((m, c))
    return m,c, error_values, mc_values
```

```python
init_m = 0.1
init_c = 0.1
learning_rate = 0.1
iterations = 250
error_threshold = 0.01
m, c, error_values, mc_values = gradient(init_m, init_c, xtrain, ytrain, learning_rate, iterations, error_threshold)
```

```python
plt.plot(np.arange(len(error_values)), error_values)
plt.ylabel('Error')
plt.xlabel('Iterations')
```

```python
predicted = (m*xtest) + c
```

```python
mean_squared_error(ytest, predicted)
```
